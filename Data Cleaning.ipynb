```python
import pandas as pd
import numpy as np
```

```python
wb = pd.read_excel('datasets/58.xls')
wb.drop(columns=['序号','标题','区','来源'],inplace = True)
wb.head()
```

```python
bk = pd.read_excel("datasets/beiker.xlsx")
bk.drop(columns=['序号','标题'],inplace = True)
bk.rename(columns={'平米':'面积','地址':'小区'},inplace = True)
bk.head()
```

```python
jk = pd.read_excel("datasets/anjuke.xls")
jk.drop(columns=['序号','标题'],inplace = True)
jk.rename(columns={'平米':'面积'},inplace = True)
jk['地址']= jk['地址'].str.split(' ').str[57]
jk['地址']= jk['地址'].str.split('\n').str[0]
jk.head()
```

```python
lj = pd.read_excel("datasets/lianjia.xls")
lj.drop(columns=['序号','标题','区1','区2'],inplace = True)
lj['面积'], lj['面积2'] = lj['面积'].str.split('㎡').str
lj['面积2'] = lj['面积2'].str.split('南').str[1]
lj['面积2'] = lj['面积2'].str.split(' ').str[1]
lj['面积2'] = lj['面积2'].str.split('北').str[1]
lj['室'], lj['面积2'] = lj['面积2'].str.split('室').str
lj['厅'], lj['面积2'] = lj['面积2'].str.split('厅').str
lj['卫'], lj['面积2'] = lj['面积2'].str.split('卫').str
lj.drop(columns=['面积2'],inplace = True)
lj.dropna(axis=0,how='any',inplace = True)
lj.head()
```

```python
df = pd.concat([wb,bk,jk,lj],axis=0)
df['价格'] = pd.to_numeric(df['价格'], downcast='integer',errors='coerce')
df['室'] = pd.to_numeric(df['室'], downcast='integer',errors='coerce')
df['厅'] = pd.to_numeric(df['厅'], downcast='integer',errors='coerce')
df['卫'] = pd.to_numeric(df['卫'], downcast='integer',errors='coerce')
df['面积'] = pd.to_numeric(df['面积'], downcast='integer',errors='coerce')
df.head()
```
