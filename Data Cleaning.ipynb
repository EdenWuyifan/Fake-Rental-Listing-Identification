```python
import pandas as pd
import numpy as np
import math
```

```python
wb = pd.read_excel('datasets/58.xls')
wb.drop(columns=['序号','标题','区','来源'],inplace = True)

biaoqians = wb["标签"]
subways = []

for tag in biaoqians:
    subway = 0
    try:
        m = tag.find("m")
        x = tag.find("行")
        distance = int(tag[x+1:m])
        if distance <= 1000:
            subway = 1
    except:
        subway = 0
    subways.append(subway)
wb.drop(columns=['标签'],inplace = True)
wb["近地铁"] = subways

wb.head()
```

```python
xlsx = pd.ExcelFile("datasets/beiker.xlsx")
bk = pd.read_excel(xlsx, 0)
df1 = pd.read_excel(xlsx, 1)
df2 = pd.read_excel(xlsx, 2)
df3 = pd.read_excel(xlsx, 3)
bk = bk.append(df1)
bk = bk.append(df2)
bk = bk.append(df3)
bk.drop(columns=['序号','标题'],inplace = True)
bk.rename(columns={'平米':'面积','地址':'小区'},inplace = True)

biaoqians = bk["标签"]
subways = []
looks = []
delicates = []
for tag in biaoqians:
    subway = 0
    look = 0
    delicate = 0
    try:
        if "近地铁" in tag: subway = 1
        if "精装" in tag: delicate = 1
        if "随时看房" in tag: look = 1
    except:
        pass
    subways.append(subway)
    looks.append(look)
    delicates.append(delicate)

bk.drop(columns=['标签'],inplace = True)
bk["近地铁"] = subways
bk["随时看房"] = looks
bk["精装"] = delicates
bk.head()
```

```python
jk = pd.read_excel("datasets/anjuke.xls")
jk.drop(columns=['序号','标题','地址'],inplace = True)
jk.rename(columns={'平米':'面积'},inplace = True)
#jk['地址']= jk['地址'].str.split(' ').str[57]
#jk['地址']= jk['地址'].str.split('\n').str[0]

floors = jk["楼层"]
floors_new = []
for floor in floors:
    fl = 0
    try:
        tol = floor.find("(")
        ed = floor.find(")")
        tol = int(floor[tol+2:ed-1])
        if floor[:2] == "中层": fl = int(math.ceil(tol/2))
        elif floor[:2] == "高层": fl = int(math.ceil(tol*3/4))
        else: fl = int(math.ceil(tol*1/4))
    except:
        tol = floor.find("共")
        ed = floor.find("层")
        tol = int(floor[tol+1:ed])
        fl = fl = int(math.ceil(tol/2))
    floors_new.append(fl)

jk["楼层"] = floors_new   

biaoqians = jk["标签"]
subways = []
for tag in biaoqians:
    subway = 0
    try:
        if "号线" in tag: subway = 1
    except:
        pass
    subways.append(subway)


jk.drop(columns=['标签'],inplace = True)
jk["近地铁"] = subways
        
jk[:20]
```

```python
lj = pd.read_excel("datasets/lianjia.xls")
lj.drop(columns=['序号','标题','区1','区2'],inplace = True)
lj['面积'], lj['面积2'] = lj['面积'].str.split('㎡').str
lj['面积2'] = lj['面积2'].str.split('南').str[1]
lj['面积2'] = lj['面积2'].str.split(' ').str[1]
lj['面积2'] = lj['面积2'].str.split('北').str[1]
lj['室'], lj['面积2'] = lj['面积2'].str.split('室').str
lj['厅'], lj['面积2'] = lj['面积2'].str.split('厅').str
lj['卫'], lj['面积2'] = lj['面积2'].str.split('卫').str
lj.drop(columns=['面积2'],inplace = True)
lj.dropna(axis=0,how='any',inplace = True)
lj.head()
```

```python
zr = pd.read_excel("datasets/ziroom.xls")

zr['楼层'] = zr['楼层'].str.split('/').str[0]

zr = zr.rename(columns={'标题': '小区'})

xiaoqus = zr['小区']
xiaoqus_new = []
for xiaoqu in xiaoqus:
    dot = xiaoqu.find("·")
    shi = xiaoqu.find("室")
    xiaoqus_new.append(xiaoqu[dot+1:shi-1])
zr['小区'] = xiaoqus_new


dizhis = zr["地址"]
subways = []

for tag in dizhis:
    subway = 0
    try:
        m = tag.find("米")
        x = tag.find("约")
        distance = int(tag[x+1:m])
        if distance <= 1000:
            subway = 1
    except:
        pass
    subways.append(subway)

zr["近地铁"] = subways
zr.drop(columns=['标签'],inplace = True)
zr.drop(columns=['地址'],inplace = True)
zr.head()
```

```python
df = pd.concat([wb,bk,jk,lj,zr],axis=0)
df['价格'] = pd.to_numeric(df['价格'], downcast='integer',errors='coerce')
df['室'] = pd.to_numeric(df['室'], downcast='integer',errors='coerce')
df['厅'] = pd.to_numeric(df['厅'], downcast='integer',errors='coerce')
df['卫'] = pd.to_numeric(df['卫'], downcast='integer',errors='coerce')
df['面积'] = pd.to_numeric(df['面积'], downcast='integer',errors='coerce')

df = df[['小区','价格','面积','室','厅','卫','楼层','近地铁','随时看房','精装']]
df.head()
```

```python
df.to_excel('data_after_clean.xls')
```
