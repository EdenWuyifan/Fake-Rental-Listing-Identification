```python
from urllib import request
import re
import pandas as pd
import urllib.parse as urp
import requests
import json

import csv
```

```python
def __get_location1__(name,city):  
        my_ak = 'uXuYBYbI90SdKOGSjIqgDnthz6DrkFN4'   
        qurey = urp.quote(name)
        try:
            url = 'http://api.map.baidu.com/place/v2/search?query='+qurey+'&tag='+'&region='+urp.quote(city)+'&output=json&ak='+my_ak
            #print(url)
            req = request.urlopen(url)
            res = req.read().decode()
            temp = json.loads(res)
            lat = pd.to_numeric(re.findall('"lat":(.*)',res)[0].split(',')[0])
            lng = pd.to_numeric(re.findall('"lng":(.*)',res)[0])
            area = re.findall('"area":(.*)',res)[0][:-1]
            return (lng,lat),str(area)  #经度和纬度
        except:
            return 0,0,''
```

# 58同城

```python
df = pd.read_excel('datasets/58.xls')
df=df.drop(columns=["区","来源","标题","序号","标签"])
sample = df
sample.head()
```

```python
sample['经度']=0
sample['纬度']=0
sample['area']=0
sample = sample[['经度','纬度','室','厅','面积','价格','小区','area']]
sample.head()
#sample['黄浦区']=0
#sample['徐汇区']=0
#sample['长宁区']=0
#sample['静安区']=0
#sample['普陀区']=0
#sample['虹口区']=0
#sample['杨浦区']=0
#sample['闵行区']=0
#sample['宝山区']=0
#sample['嘉定区']=0
#sample['金山区']=0
#sample['松江区']=0
#sample['青浦区']=0
#sample['奉贤区']=0
#sample['崇明区']=0
#sample['浦东新区']=0
#sample = sample[['经度','纬度','室','厅','面积','价格','小区','黄浦区','徐汇区','长宁区','静安区','普陀区','虹口区','杨浦区','闵行区','宝山区','嘉定区','金山区','松江区','青浦区','奉贤区','崇明区','浦东新区']]
```

```python
sample['经度']=0
sample['纬度']=0

location = sample['小区']
lng = []
lat = []
area_lst = []

for loc in location:
    result = __get_location1__(str(loc),'上海')
    address=result[0]
    area=result[1]
    
    while address == (0,0):
        for i in range(10):
            address = __get_location1__(str(loc),'上海')
            if address ==(0,0):
                continue
            else:
                break
        break
    
    lng.append(str(address[0]))
    lat.append(str(address[1]))
    area_lst.append(area)

sample['经度']=lng
sample['纬度']=lat
sample['area']=area_lst
dummy = pd.get_dummies(sample['area'])
merged = pd.concat([sample,dummy], axis=1)
merged = merged.drop(columns="area")
merged.head()
```

```python
merged = merged.drop(columns=["小区"])
merged.to_excel('58_qu.xls')
```

```python
sample = sample.drop(columns=["小区"])
sample.to_excel('58_new.xls')
```

# Lianjia

```python
xlsx = pd.ExcelFile("datasets/lianjia.xls")
df = pd.read_excel(xlsx, 0)
df = df.drop(columns=["序号","标题"])
df.head()
```

```python
area = []
shi = []
ting = []
for i in df['面积']:
    a = i.find('㎡')
    s = i.find('室')
    t = i.find('厅')
    area.append(i[:a])
    shi.append(i[s-1:s])
    ting.append(i[t-1:t])
df['面积']=area
df['室']=shi
df['厅']=ting
df['经度']=0
df['纬度']=0

df = df[['经度','纬度','室','厅','面积','价格','小区']]
df.head()
```

```python
location = df['小区']
lng = []
lat = []

for loc in location:
    address=__get_location1__(str(loc),'上海')
    
    while address == (0,0):
        for i in range(5):
            address = __get_location1__(str(loc),'上海')
            if address ==(0,0):
                continue
            else:
                break
        break
    
    lng.append(str(address[0]))
    lat.append(str(address[1]))

df['经度']=lng
df['纬度']=lat

df.head()
```

```python
df=df.drop(columns=["小区"])
df.to_excel('lianjia_new.xls')
```
